```python
from nltk import sent_tokenize
from nltk import word_tokenize
from nltk.corpus import stopwords
```

```python
text = "The Geese were flying dangerously and eating the best fishes. The kids were playing football and a coast guard was looking at the shores."
```

```python
tokens_sents = sent_tokenize(text)
print("Sentences:\n", tokens_sents)
```

```python
token_words = word_tokenize(text)
print("\nWords:\n", token_words)
```

```python
from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import LancasterStemmer
```

```python
stem = []

for i in token_words:
    ps = PorterStemmer()
    stem_word = ps.stem(i)
    stem.append(stem_word)

print(stem)
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
tagged_token = nltk.pos_tag(token_words)
print("Parts of Speech: ", tagged_token)
```

```python
from nltk.corpus import wordnet

def get_word_POS(tag):
    if tag.startswith("J"):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('R'):
        return wordnet.ADV
    elif tag.startswith('N'):
        return wordnet.NOUN
    else:
        return wordnet.NOUN
```

```python
leme = []

for i, w_tag in tagged_token:
    word = lemmatizer.lemmatize(i.lower(), pos=get_word_POS(w_tag))
    leme.append(word)

print(leme)
```

```python
lemmatized_output = ' '.join([
    lemmatizer.lemmatize(w.lower(), pos=get_word_POS(tag))
    for w, tag in tagged_token
])

print(lemmatized_output)
```
